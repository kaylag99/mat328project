# Data Cleaning

```python
import folium
import pandas as pd
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.formula.api as smf
import seaborn as sns
import numpy as np
```

```python
crash = pd.read_csv("Motor_Vehicle_Collisions_-_Crashes.csv", na_values = ["Unspecified"])
pd.set_option('display.max_columns', None)
crash.head()
```

```python
new_columns = ["date","time","borough","zipcode", "latitude", "longitude", "location", "on_street", "cross_street", "off_street", "num_persons_injured", "num_persons_killed", "num_ped_injured", "num_ped_killed", "num_cyc_injured", "num_cyc_killed", "num_mot_injured", "num_mot_killed", "con_vehicle_1", "con_vehicle_2", "con_vehicle_3", "con_vehicle_4", "con_vehicle_5", "collision_id", "vehicle_type_1", "vehicle_type_2", "vehicle_type_3", "vehicle_type_4", "vehicle_type_5"]
crash.columns = new_columns
```

```python
crash["date"] = pd.to_datetime(crash["date"])
crash.head()
```

```python
crash["date"] = crash["date"].dt.year
```

```python
crash = crash[["date","latitude", "longitude"]]
```

# Maps

```python
filt1 = crash["date"] == 2020
crash2020 = crash[filt1]
crash2020.head()
```

```python
crash2020.loc[50:55]
```

```python
crash2020.reset_index(inplace = True, drop = True)
```

```python
len(crash2020)
```

```python
sample100 = crash2020.sample(100)
```

```python
sample100 = sample100.dropna()
```

```python
sample100.reset_index(inplace = True, drop = True)
```

```python
crash_map = folium.Map(location=[40.7128, -74.0060])
crash_map
```

```python
for i in range(len(sample100)):
    row = sample100.loc[i]
    print("Coordinates: ", row["latitude"], row["longitude"] )
```

```python
crash_map = folium.Map(location=[40.7128, -74.0060])
for i in range(len(sample100)):
    row = sample100.loc[i]
    folium.Marker([row["latitude"], row["longitude"]], 
                 # tooltip=row["on_street"]
             ).add_to(crash_map)
```

Generating this map took multiple steps. In addition to the steps I completed in the initial cleaning, I filtered the data so that I only have car crashes that happened in 2020. Additionally, I sliced that dataset so that I only had three columns, date, latitude, and longitude. I then took a sample of 100 from the 112,000 collisions that occurred last year because it was impossible to plot all those points on this map. 

The first step I took to create the maps was to create an empty map of New York City. Next, I used iteration, or loops, to loop through all the rows of the dataset, store the rows in the variable "row", and used it to print the latitude and longitude. Lastly, I plotted the crashes on the map. The latitudes and longitudes from each row were used to create all the markers for my map. 

Pictured below/above is a map of New York City. Plotted on the map are 100 of the 112,834 collisions that occurred in 2020. There are 20 in the Bronx, 14 in Manhattan, and 2 in Staten Island. Between Brooklyn and Manhattan, there were 64 crashes (I cannot tell where the boundaries for the two boroughs are).
